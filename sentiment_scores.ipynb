```python
from allennlp.predictors.predictor import Predictor
import allennlp_models.tagging

#imported predictor
predictor = Predictor.from_path("https://storage.googleapis.com/allennlp-public-models/basic_stanford_sentiment_treebank-2020.06.09.tar.gz")
```

```python
import pandas as pd 
import pickle

#imported file from drive
infile = open(r'/content/drive/MyDrive/data/complete.pickle','rb')
new_dict = pickle.load(infile)
infile.close()
#all categories of data
print((new_dict.keys()))

#made a dataframe
df=pd.DataFrame.from_dict(new_dict['tweet_raw_text'])
```

```python
df.iloc[2].values[0]
```

```python
length = df.shape[0]
#containing the dictionary of results 
dict_results=[]
print(length)
for i in range(length):
  #print(str(df.iloc[i].values[0]))
  text=str(df.iloc[i].values[0])
  try:
    #appending in list
    temp=predictor.predict(text)
    dict_results.append((i ,temp['probs'][1]))
  except:
    #empty text section 
    print(i)
  

```

```python
#final tweets with scores
print(len(dict_results))

#loaded to drive in a pickle
with open('/content/drive/MyDrive/data/anubhav.pickle', 'wb') as handle:
    pickle.dump(dict_results , handle, protocol=pickle.HIGHEST_PROTOCOL)
```
